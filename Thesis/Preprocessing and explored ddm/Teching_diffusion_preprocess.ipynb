###Analyze Key Categories

```python
# distribution of diagram types

import seaborn as sns
import matplotlib.pyplot as plt

# Get the counts of each diagram type
diagram_type_counts = df['Diagram Type'].value_counts()
print("Distribution of Diagram Types:")
print(diagram_type_counts)

# --- Visualize the distribution ---
plt.figure(figsize=(10, 6))
sns.barplot(x=diagram_type_counts.values, y=diagram_type_counts.index, orient='h')
plt.title('Number of Examples per Diagram Type')
plt.xlabel('Count')
plt.ylabel('Diagram Type')
plt.show()
```

```python
# Get the counts of each level
level_counts = df['Level'].value_counts()
print("\nDistribution of Levels:")
print(level_counts)

# --- Visualize the distribution ---
plt.figure(figsize=(8, 5))
sns.barplot(x=level_counts.index, y=level_counts.values)
plt.title('Number of Examples per Level')
plt.xlabel('Level')
plt.ylabel('Count')
plt.show()
```
